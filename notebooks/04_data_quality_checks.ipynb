# 04 — Data Quality Checks

**Week:** 6

Goal: Run meaningful data quality rules and document failures.


```python
silver_df = spark.table("silver_standardized_events")
silver_df.createOrReplaceTempView("silver_events")
```

```python
%sql
-- DQ-01: Required record_id should not be null
SELECT COUNT(*) AS failed_count
FROM silver_events
WHERE record_id IS NULL
```

```python
%sql
-- DQ-02: Duplicate record_id check
SELECT record_id, COUNT(*) AS duplicate_count
FROM silver_events
GROUP BY record_id
HAVING COUNT(*) > 1
```

```python
%sql
-- DQ-03: Amount should not be negative
SELECT *
FROM silver_events
WHERE amount < 0
LIMIT 20
```

```python
%sql
-- DQ-04: Event timestamp should not be null
SELECT COUNT(*) AS failed_count
FROM silver_events
WHERE event_timestamp IS NULL
```
