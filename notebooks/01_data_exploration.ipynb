Goal: Load raw files in Databricks, inspect schemas, count rows, identify nulls and profile early issues.


```python
# Update paths based on your Databricks upload location.
# Example volume path:
# raw_path = "/Volumes/workspace/default/<project_name>/raw/source_events_raw.csv"

raw_path = "/Volumes/workspace/default/<project_name>/raw/source_events_raw.csv"
df = spark.read.option("header", True).option("inferSchema", True).csv(raw_path)
display(df.limit(10))
```

```python
df.printSchema()
print("Row count:", df.count())
```

```python
# Register temporary view for SQL-first exploration
df.createOrReplaceTempView("raw_source_events")
```

```python
%sql
SELECT status, COUNT(*) AS record_count
FROM raw_source_events
GROUP BY status
ORDER BY record_count DESC
```
