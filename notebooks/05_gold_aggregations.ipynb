# 05 — Gold Aggregations

**Week:** 7

Goal: Create dashboard-ready Gold metric tables.


```python
%sql
CREATE OR REPLACE TABLE gold_event_summary_by_date AS
SELECT
  event_date,
  status_standard,
  COUNT(*) AS record_count,
  SUM(amount) AS total_amount,
  AVG(amount) AS avg_amount
FROM silver_standardized_events
GROUP BY event_date, status_standard
```

```python
%sql
SELECT *
FROM gold_event_summary_by_date
ORDER BY event_date, status_standard
```
